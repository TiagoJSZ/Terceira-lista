```python
# Método de Euler


import numpy as np
import matplotlib.pyplot as plt

plt.style.use('seaborn-poster')
%matplotlib inline

# Definição de função
f = lambda t, s: t**2 + t**(1/3)
h = 0.1 # Step size
t = np.arange(0, 0.7, h) # grid numérico
s0 = 7 # condição inicial

# Método de Euler
s = np.zeros(len(t))
s[0] = s0

for i in range(0, len(t) - 1):
    s[i + 1] = s[i] + h*f(t[i], s[i])

plt.figure(figsize = (12, 8))
plt.plot(t, s, 'bo--', label='Aproximação')
plt.plot(t, -np.exp(-t), 'g', label='Exato')
plt.title('Solução aproximada e exata')
plt.xlabel('t')
plt.ylabel('f(t)')
plt.grid()
plt.legend(loc='lower right')
plt.show()
```
