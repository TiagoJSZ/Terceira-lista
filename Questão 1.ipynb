```python
import numpy as np
import matplotlib.pyplot as plt
nu=1e12*np.array([1.44,1.33,1.22,1.10,0.95],dtype=float)
Vp=np.array([5.76,5.3361,4.9284,4.4944,3.8025])
g=9.8
Vp=2*np.pi*(nu/g)
Sxx=0;Sxy=0
N=len(nu)
for k in range(N):
    Sxy=Sxy+nu[k]*Vp[k]
    Sxx=Sxx+nu[k]*nu[k]
    Sx=np.sum(nu)
    Sy=np.sum(Vp)
    a=(Sxy-Sx*Sy/N)/(Sxx-Sx**2/N)
    b=(Sy-a*Sx)/N
h=a
W=b
plt.close('all')
plt.scatter(nu,Vp)
fit=h*nu+W;
plt.plot(nu,fit)
plt.xlabel(r'$\nu$')
plt.ylabel(r'$V_p$')
plt.title('Determinação da constante de Planck')
```
